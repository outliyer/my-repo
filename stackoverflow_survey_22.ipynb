```python
india=df[df['Country']=='India']
india['EdLevel'].dropna(inplace=True)
```

# insigts from india

```python
# visualizing the education level in indial
plt.style.use('ggplot')
plt.figure(figsize=(12,8))
sns.barplot(x=india['EdLevel'].value_counts(),y=india['EdLevel'].unique(),data=india,palette=color)
plt.title('Education level of the programmers in india')
```

```python
india['LanguageHaveWorkedWith']=india['LanguageHaveWorkedWith'].str.replace(';',',')
```

```python
india['LanguageHaveWorkedWith'].value_counts()
```

```python
list=['Python','Java','SQL','JavaScript','HTML/CSS','C','Ruby','Bash/Shell','Kotlin','Rust']
program=[]
for i in list:
    k=india['LanguageHaveWorkedWith'].str.contains(i,na=False).sum()
    program.append(k)
```

```python
plt.figure(figsize=(12,6))
# k=sns.color_palette('YlOrRd')
sns.barplot(x=list,y=program,palette=color)
plt.title("popular languages in india")
plt.style.use('ggplot')
```

```python
work_grp=india.groupby(['WorkExp'])
work_grp=work_grp['CompTotal'].mean()
```

```python
plt.figure(figsize=(12,6))
sns.distplot(india['WorkExp'])
plt.title('work experience distrubution')
```

```python
plt.figure(figsize=(12,6))
plt.title('mean_salary of a employe with work experencie')
plt.plot(work_grp)
```

```python
profession=[]
prof=[ 'full-stack','back-end','front-end','Data scientist','DevOps','Blockchain','Cloud infrastructure engineer']
for i in prof:
    k=india['DevType'].str.contains(i).sum()
    profession.append(k)

    
```

```python
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.pie(profession,labels=prof,data=india,wedgeprops={'edgecolor':'k'},radius=0.8,colors=color)
plt.pie([1],radius=0.54,colors='w')
plt.title('programmers by profession in india')
plt.style.use("seaborn")
plt.show()
```
